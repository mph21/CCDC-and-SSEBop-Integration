# Automated Raster builder for CCDC

# Build list of Dates from Landsat

```python
#read in lsatIDs
meta = pd.read_csv("D:/Users/mhiestand/Documents/ARD_CHIPS_pr/colorado_forest_2019_lsatIDS.csv")
ids = meta["ID"].to_numpy()

#Build list of dates 
dates = []

for lsatid in ids:
    tdate = datetime.datetime.strptime(lsatid[-8:], '%Y%m%d').strftime('%Y-%m-%d')
    tdate = datetime.datetime.strptime(tdate, '%Y-%m-%d')
    dates.append(tdate)

dates = sorted(dates)

print(len(ids))
print(ids)
```

# Generate Synthetic CCDC data sets

```python
import import_ipynb

from functions_synth_ndwi import super_NDWI
from functions_synth_ndwi import water_mask
from functions_synth_ndvi import super_NDVI
from functions_CCDC_SBtemp import super_grid
```

```python
#Generate Synthetic CCDC Data Sets
print("Surface Brightness")
asbt, tempx, tempy = super_grid(somex, somey, dates, numchips)
print("NDVI")
andvi, tempx, tempy = super_NDVI(somex, somey, dates, numchips)
print("NDWI")
andwi, tempx, tempy = super_NDWI(somex, somey, dates, numchips)
```

# Output multi band Rasters for SSEBop

```python
#Output file stuff
xnot = min(tempx)
ynot = max(tempy)

transform = rasterio.transform.from_origin(xnot, ynot, 30, 30)  
root = 'tenChips_Colorado_Forest/forestCO_'

count = 0

for n in enumerate(ids):
    
    # put each layer of array into output list
    nbt = asbt[:,:,count]
    nandvi = andvi[:,:,count]
    nandwi = andwi[:,:,count]
    water2 = water_mask(nandwi, 0.1)
    outstack = [nbt, nandvi, nandwi, water2]
    

    #output file meta data
    lsatid = ids[count]
    path = root + lsatid[-15:] + '.tif'
    new_dataset = rasterio.open(path, 'w', driver='GTiff',
                            height = nbt.shape[0], width = nbt.shape[1],
                            count=4, dtype=str(nbt.dtype),
                            crs=5070,
                            transform=transform)


    #output as multiband raster
    for band_nr, src in enumerate(outstack, start=1):
        new_dataset.write(src, band_nr)


    new_dataset.close()
    print(path)
    count = count+1

```
